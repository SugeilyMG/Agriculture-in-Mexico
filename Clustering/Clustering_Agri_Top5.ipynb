# Preprocessing

```python
# Convert Categorical Varibales into Numeric Variables
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
# Check the new size of the df
X.shape
```

## Scaling

```python
# Start the process of scaling your data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

## PCA

```python
# Import PCA
from sklearn.decomposition import PCA 
```

```python
# Load PCA and pass the scaled data into it
pca = PCA()
pca.fit_transform(scaled_X)
```

```python
len(pca.explained_variance_ratio_)
```

```python
# Plot the principal components with its variance 
plt.figure(figsize=(10,8))
plt.plot(range(1,18), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
# We use 8 components since it explains around 90% of our dataset variance
pca = PCA(n_components=8)
```

```python
# Repeat PCA process
pca.fit(scaled_X)
```

```python
pca.transform(scaled_X)
```

```python
scores_pca = pca.transform(scaled_X)
```

# Clustering 

```python
# Import KMeans to do the clusters
from sklearn.cluster import KMeans
```

```python
# Start getting the within clusters sum of squared distances by testing 2 to 20 clusters
wcss = []

for k in range(2,20):
    kmeans_pca = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
# Create elbow plot
plt.plot(range(2,20), wcss, 'o--')
plt.xlabel('Number of Clusters')
plt.ylabel('Within Clusters Sum of Squared Distances')
plt.title('K-means with PCA Clustering')
```

```python
# Create bar plot with to see the differences between the sum of squared distances
pd.Series(wcss).diff().plot(kind='bar')
```

```python
# By checking the plot of above we decide to select 8 as our number of clusters
model = KMeans(n_clusters=8)
model.fit(scaled_X)
```

```python
# Get the clust labels
model.labels_
```

```python
# Add the cluster labels into your dataset
X['K=8 Cluster Labels'] = model.labels_
```

```python
# Verify the change was made
X.head()
```
